```r
library(phyloseq)
library(tibble)
library(dplyr)
```

```r
asv_sheet <- read.csv("asv_sheet.csv")
colnames(asv_sheet)[1] <- "Row_ID"
asv_sheet <- tibble::column_to_rownames(asv_sheet, "Row_ID") 
```

```r
#2.2 Import taxa table
tax_sheet_raw <- read.csv("tax_sheet.csv", row.names=1)
str(tax_sheet_raw)
tax_sheet <- subset(tax_sheet_raw, select = c("superkingdom", "phylum", "class", "order","family", "genus", "species"))
colnames(tax_sheet) <- c("Domain", "Phylum", "Class", "Order", "Family", "Genus", "Species")
str(tax_sheet)
```

```r
#2.3 Import sample sheet table
# We have to make the char columns factors to avoid downstream issues
sample_sheet <- read.csv("sample_sheet.csv")  %>%
                tibble::column_to_rownames(var = "Row_ID") %>%
                dplyr::mutate_if(is.character, as.factor)
str(sample_sheet)                
```

```r
sample_sheet
```

```r
# Get an idea of how to filter our data.frame columns of interest
sample_sheet %>%
dplyr::select(Site, sample_nc) %>%
dplyr::distinct()
```

```r
# Planning for the pooling 
# Because we are only going to group some, and not all samples, we have to introduce a grouping variable that will group only our sample of interest
# We are going to pool the field blanks by site, where expect to pool the six sample into two samples making the total per site be 15 + 1 pooled field blank
```

```r
#We create a new sample sheet check the sample names in Sample Name2
(sample_sheet2 <- sample_sheet %>%
                  tibble::rownames_to_column(var = "Row_ID")%>%
                  dplyr::group_by(Site, sample_nc) %>%
                  #create a new column for store group name
                  dplyr::mutate(group_name = case_when(Site == "sea" & sample_nc == "field_nc" ~ "sea_field_nc",
                                                  Site ==  "pond" & sample_nc == "field_nc" ~ "pond_field_nc")) %>%
                  dplyr::ungroup() %>%
                  #once we have the grouping names we going to use them as sample names
                  dplyr::mutate(Sample_Name2 = factor(dplyr::if_else(is.na(group_name), Sample_Name2, group_name)))%>%
                  tibble::column_to_rownames("Row_ID"))
```

```r
#create a phyloseq sample_data object using our new samplesheet2
SAMPLE_DATA <- sample_data(sample_sheet2)

#set taxa are rows to FALSE to avoid downstream errors
OTU = otu_table(as.matrix(asv_sheet), taxa_are_rows = FALSE)

TAX = tax_table(as.matrix(tax_sheet))

#Sanity check if have formatted the table correctly and everything aligns before creating our phyloseq object
all(taxa_names(TAX) == taxa_names(OTU))
```

```r
(physeq <- phyloseq(OTU, TAX, SAMPLE_DATA))
```

```r
# Add a variable "substrate type" to the sample data table, by adding a column with two levels, sediment and water 
# Create a new sample variable named "substrate" with appropriate levels
substrate_type <- c(rep("sediment", 18), rep("water", 18))
substrate_df <- data.frame(sample_names = sample_names(physeq), substrate_type) %>%
                tibble::column_to_rownames(var = "sample_names")

# Create a phyloseq sample data object
substrate_sample_data <- phyloseq::sample_data(substrate_df)

# Merge the new sample data into the existing phyloseq object
# Check if there is a a change in number of columns between old and merged phylose object
(physeq_new <- merge_phyloseq(physeq, substrate_sample_data))
# It was not clear to me how you planned to use this substrate data 
# I will stop here and continue with the main issue of grouping select samples
```

```r
# Check for counts 
# Remember the order of variables changes matters here
new_sample_data <- sample_data(physeq) %>% data.frame()
xtabs(~sample_nc + Site, data = new_sample_data)
```

```r
# We can merge based on SampleName2 as they are unique and contain our grouping 
(merged_physeq <- phyloseq::merge_samples(physeq, "Sample_Name2"))
# This generates some warning messages because phylose does not know how the merging the colums with characters
```

```r
# This works well for the otu table
otu_table(merged_physeq)
```
